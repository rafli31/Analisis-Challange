```python
# Import data
import pandas as pd
Teams = pd.read_csv('https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/Teams.csv')
Teams.head()
```

### 1. Team (teamID )dengan pertandingan home terendah

#### Cara Cepat

```python
# Mencari TeamID dengan pertandingan home terendah
df_sorted = Teams.sort_values(by='Ghome')
df_sorted
```

#### Mencari team ID dengan pertadingan home terendah

Menggunakan Variabel Berikut : <br>
- teamID -> 	Team; a factor <br>
- Ghome  ->   Games played at home

```python
# Memilih variabel team ID dan Ghome
Data_home = Teams[['teamID', 'Ghome']]

# Membuat DataFrame dari data
df_home = pd.DataFrame(Data_home)
df_home
```

```python
## Menghilangkan missing Value

# Menghapus baris yang berisi nilai yang hilang
df_cleaned = df_home.dropna()

# Hasil setelah menghilangkan missing value
data_pertandingan_home = pd.DataFrame(df_cleaned)
data_pertandingan_home
```

```python
# Mengambil data teamID 
data_pertandingan_home['teamID'].unique()
```

```python
# Memfilter Team (teamID ) dengan pertandingan home terendah
data_pertandingan_home.sort_values(by='Ghome')
```

### Interpretasi

Berdasarkan hasil yang ditampilkan, dapat diketahui bahwa Team (teamID ) dengan pertandingan home terendah yaitu dengan teamID **"SEA"** dengan nilai Ghome sebesar **24.0**

### 2. Nama Team (name) dan Tahun (yearID) dimana Kemenangan (W) terbanyak terjadi.

Menggunakan Variabel Berikut : <br>
- name	    -> Teamâ€™s full name <br>
- yearID	-> Year <br>
- W	        -> Wins

```python
# Mengelompokan variabel yang dibutuhkan
data_team = Teams[['name', 'yearID', 'W']]
df_team = pd.DataFrame(data_team)
df_team
```

```python
# Mengelompokkan data berdasarkan 'name' dan 'yearID', kemudian menjumlahkan kemenangan (W)
grouped = df_team.groupby(['name', 'yearID'])['W'].sum().reset_index()
grouped
```

```python
# Menemukan baris dengan kemenangan terbanyak
baris_terbanyak = grouped[grouped['W'] == grouped['W'].max()]
baris_terbanyak
```

```python
# Menampilkan hasil
print("Nama Team (name) dengan Kemenangan (win) terbanyak:", baris_terbanyak['name'].values[0])
print("Tahun (yearID) dimana Kemenangan (win) terbanyak terjadi:", baris_terbanyak['yearID'].values[0])
print("Jumlah Kemenangan terbanyak:", baris_terbanyak['W'].values[0])
```

#### Interpretasi

Berdasarkan hasil di atas, dapat diketahui bahwa Nama Team (name) dan Tahun (yearID) dimana Kemenangan (W) terbanyak terjadi yaitu **New York Giants** yang terjadi pada **tahun 1890** dengan kemenangan sebanyak **137 kemenangan**.

### 3 Tahun (yearID) dimana pertandingan terbanyak (G) terjadi, serta jumlah pertandingannya

Menggunakan Variabel Berikut : <br>
- yearID ->	Year <br>
- G	     -> Games played

```python
# Mengelompokan variabel yang dibutuhkan
data_pertandingan = Teams[['yearID', 'G']]
df_pertandingan = pd.DataFrame(data_pertandingan)
df_pertandingan
```

```python
# Mengelompokkan data berdasarkan 'tahun', kemudian menjumlahkan pertandingan
grouped_match = df_pertandingan.groupby('yearID')['G'].sum().reset_index()
grouped_match
```

```python
# Menemukan tahun dengan pertandingan terbanyak
tahun_terbanyak = grouped_match[grouped_match['G'] == grouped_match['G'].max()]
tahun_terbanyak
```

```python
# Menampilkan hasil
print("Tahun (yearID) dengan pertandingan terbanyak:", tahun_terbanyak['yearID'].values[0])
print("Jumlah pertandingan terbanyak:", tahun_terbanyak['G'].values[0])
```

### Interpretasi

Berdasarkan hasil di atas, dapat diketahui bahwa Tahun (yearID) dimana pertandingan terbanyak (G) terjadi, serta jumlah pertandingannya yaitu pada **tahun 1998** dengan jumlah pertandingannya **sebanyak 4864**.

### 4. Stadion (park) dengan total kehadiran (attendance) tertinggi dalam 5 tahun terakhir.

Menggunakan Variabel Berikut : <br>
- park	     -> Name of teamâ€™s home ballpark <br>
- yearID     ->	Year <br>
- attendance ->	Home attendance total

```python
# Mengelompokan variabel yang dibutuhkan
data_stadion = Teams[['park', 'attendance', 'yearID']]
df_stadion = pd.DataFrame(data_stadion)
df_stadion
```

```python
# Mengelompokan stadion 5 tahun terakhir
tahun_terakhir = df_stadion[df_stadion['yearID'] >= 2018]
tahun_terakhir
```

```python
# Mengelompokkan data berdasarkan 'nama_std' dan menjumlahkan pertandingan
grouped_std = tahun_terakhir.groupby('park')['attendance'].sum().reset_index()
grouped_std
```

```python
# Menemukan stadion dengan total kehadiran tertinggi
stadion_tertinggi = grouped_std[grouped_std['attendance'] == grouped_std['attendance'].max()]
stadion_tertinggi
```

```python
# Menampilkan hasil
print("Stadion (park) dengan total kehadiran tertinggi dalam 2 tahun terakhir:", stadion_tertinggi['park'].values[0])
print("Total kehadiran tertinggi:", stadion_tertinggi['attendance'].values[0])
```

### Interpretasi

Berdasarkan hasil diatas, dapat diketahui bahwa Stadion (park) dengan total kehadiran (attendance) tertinggi dalam 5 tahun terakhir adalah **Dodger Stadium** dengan **jumlah kehadirannya yaitu 14497910.0**

### 5 Tiga Nama Team (name) dengan Rata-Rata Homeruns(HRA) tertinggi dalam 10 tahun terakhir.

Menggunakan Variabel Berikut : <br>
- name	 -> Teamâ€™s full name <br>
- yearID -> Year <br>
- HRA	 -> Homeruns allowed 

```python
# Mengelompokan variabel yang dibutuhkan
data_HRA = Teams[['name', 'HRA', 'yearID']]
df_HRA = pd.DataFrame(data_HRA)
df_HRA
```

```python
# Mengambil data hanya untuk 10 tahun terakhir (tahun 2013-2022)
tahun_terakhir = df_HRA[df_HRA['yearID'] >= 2013]
tahun_terakhir
```

```python
# Mengelompokkan data berdasarkan 'team' dan menghitung rata-rata HRA
grouped_HRA = tahun_terakhir.groupby('name')['HRA'].mean().reset_index()
grouped_HRA
```

```python
# Mengurutkan DataFrame berdasarkan rata-rata HRA secara menurun
sorted_df = grouped_HRA.sort_values(by='HRA', ascending=False)
sorted_df
```

```python
# Mengambil tiga nama tim dengan rata-rata HRA tertinggi
tiga_tertinggi = sorted_df.head(3)
tiga_tertinggi
```

```python
print("Tiga Nama Team (name) dengan Rata-Rata Homeruns (HRA) tertinggi dalam 10 tahun terakhir:")
print(tiga_tertinggi)
```

### Interpretasi

Berdasarkan hasil di atas, dapat diketahui bahwa Tiga Nama Team (name) dengan Rata-Rata Homeruns(HRA) tertinggi dalam 10 tahun terakhir yaitu <br>
**1. Baltimore Orioles dengan HRA 199.9,** <br> 
**2. Cincinnati Reds dengan HRA  194.4,** <br>
**3. Seattle Mariners dengan HRA 185.9.**

### BONUS 

Seorang pemain bernama Ben bergabung dengan tim New York Mets selama tahun 2004 hingga 2012, cobalah untuk meneliti bagaimana performa tim pada periode tersebut, bandingkan dengan periode yang sama sebelum dan sesudahnya. Apakah Rekrutmen Ben merupakan hal yang positif pada Tim tersebut?

Pada kasus ini Menggunakan data <br>
- name	-> Teamâ€™s full name <br>
- yearID ->	Year
- W	    -> Wins <br>
- L	    -> Losses <br>
- LgWin	-> League Champion(Y or N)

Win dan Losses yang digunakan adalah rata-rata kemenangan dan kekalahan dari tim New York Mets selama tahun 2004 hingga 2012 dan dibandingkan sebelum dan setelah Ben bergabung, Karena hal tersebut menunjukan performa tim selama ben bergabung apakah berdampak terhadap pertandingan yang dilakukan oleh tim. <br>

Untuk LgWin digunakan jumlah juara yang diperoleh selama Ben bergabung di tim New York Mets selama tahun 2004 hingga 2012 dan dibandingkan sebelum dan setelah Ben bergabung. Hal tersebut menunjukan performa tim secara keseluruhan dengan mendapatkan trophy di akhir liga. 

```python
# Mengelompokan hanya variabel yang diinginkan
data_bonus = Teams[['name', 'W', 'L', 'yearID', 'LgWin']]
df_bonus = pd.DataFrame(data_bonus)
df_bonus
```

```python
# Memfilter data untuk tim dan tahun 2004-2012
data_tim_2004_2012 = df_bonus[(df_bonus['name'] == 'New York Mets') & (df_bonus['yearID'] >= 2004) & (df_bonus['yearID'] <= 2012)]
data_tim_2004_2012
```

```python
# Menghitung jumlah kemenangan selama 2004-2012
jumlah_win_2004_2012 = data_tim_2004_2012['W'].sum()
jumlah_win_2004_2012
```

```python
# Menghitung jumlah kekalahan selama 2004-2012
jumlah_loss_2004_2012 = data_tim_2004_2012['L'].sum()
jumlah_loss_2004_2012
```

```python
# Menghitung rata-rata kemenangan  2004-2012
rata_rata_win_2004_2012 = data_tim_2004_2012['W'].mean()
rata_rata_win_2004_2012
```

```python
# Menghitung rata-rata kekalahan  2004-2012
rata_rata_loss_2004_2012 = data_tim_2004_2012['L'].mean()
rata_rata_loss_2004_2012
```

```python
# Menghitung jumlah berapa kali tim menjadi juara 
jumlah_juara = len(data_tim_2004_2012[(data_tim_2004_2012['name'] == 'New York Mets') & (data_tim_2004_2012['LgWin'] == 'Y')])
jumlah_juara
```

```python
# Memfilter data pada tahun sebelum 2004
data_sebelum_2004 = df_bonus[(df_bonus['name'] == 'New York Mets') & (df_bonus['yearID'] < 2004)]
data_sebelum_2004
```

```python
# Menghitung jumlah kemenangan sebelum 2004
jumlah_win_sebelum_2004 = data_sebelum_2004['W'].sum()
jumlah_win_sebelum_2004
```

```python
# Menghitung jumlah kekalahan sebelum 2004
jumlah_loss_sebelum_2004 = data_sebelum_2004['L'].sum()
jumlah_loss_sebelum_2004
```

```python
# Menghitung rata-rata kemenangan sebelum 2004
rata_rata_win_sebelum_2004 = data_sebelum_2004['W'].mean()
rata_rata_win_sebelum_2004
```

```python
# Menghitung rata-rata kekalahan sebelum 2004
rata_rata_loss_sebelum_2004 = data_sebelum_2004['L'].mean()
rata_rata_loss_sebelum_2004
```

```python
# Menghitung jumlah kali tim  menjadi juara sebelum 2004
jumlah_juara2 = len(data_sebelum_2004[(data_sebelum_2004['name'] == 'New York Mets') & (data_sebelum_2004['LgWin'] == 'Y')])
jumlah_juara2
```

```python
# Memfilter data pada tahun setelah 2012
data_setelah_2012 = df_bonus[(df_bonus['name'] == 'New York Mets') & (df_bonus['yearID'] > 2012)]
data_setelah_2012
```

```python
# Menghitung jumlah kemenangan setelah 2012
jumlah_win_setelah_2012 = data_setelah_2012['W'].sum()
jumlah_win_setelah_2012
```

```python
# Menghitung jumlah kekalahan setelah_2012
jumlah_loss_setelah_2012= data_setelah_2012['L'].sum()
jumlah_loss_setelah_2012
```

```python
# Menghitung rata-rata kekalahan setelah_2012
rata_rata_loss_setelah_2012 = data_setelah_2012['L'].mean()
rata_rata_loss_setelah_2012
```

```python
# Menghitung jumlah kali tim  menjadi juara sebelum 2004
jumlah_juara3 = len(data_setelah_2012[(data_setelah_2012['name'] == 'New York Mets') & (data_setelah_2012['LgWin'] == 'Y')])
jumlah_juara3
```
